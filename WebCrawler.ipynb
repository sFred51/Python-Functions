```python
#Date Created: 2015-11-04
#Last Editted: 2015-11-04
#Author(s): Steven Henkel
#Edited by: Steven Henkel
#Version: 1
#Purpose: The purpose of this function is to convert a python datetime into a recognizable format for Oracle (YYYY-MM-DD HH24:MI:SS)
#    dateTime <datetime>: The date you want to convert
def toSqlDateTime (dateTime):
    return str(dateTime.year)+"-"+zeroStrings(str(dateTime.month),2)+"-"+zeroStrings(str(dateTime.day),2)+" "+str(dateTime.hour)+":"+str(dateTime.minute)+":"+str(dateTime.second)
```
